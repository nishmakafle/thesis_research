```python
import time
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
X = np.linspace(0, 10, 100)
Y = 3 * X + 2 + np.random.randn(100) * 2

m = np.random.randn()
c = np.random.randn()

learning_rate = 0.01
epochs = 1000 

for i in range(epochs):
    Y_pred = m * X + c 
    error = Y_pred - Y

    dm = (2 / len(X)) * np.sum(error * X)
    dc = (2 / len(X)) * np.sum(error)

    m -= learning_rate * dm
    c -= learning_rate * dc

    if i % 50 == 0:
        loss = np.mean(error**2)
        time.sleep(1)
        print(f"Epoch {i}: Loss = {loss:.4f}, m = {m:.4f}, c = {c:.4f}")

print(f"\nFinal values: m = {m:.4f}, c = {c:.4f}")

plt.scatter(X, Y, label="Data")
plt.plot(X, m * X + c, color='red', label="Best Fit Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
